```python
import pandas as pd
import numpy as np
from scipy import stats 

def calcular_lift_generico(df_analise: pd.DataFrame, coluna_metrica: str) -> dict:
    """
    Calcula o Lift (ganho percentual) e a Significância Estatística (P-valor) 
    para uma métrica contínua em um teste A/B, usando 'control' e 'target' como rótulos.

    Args:
        df_analise (pd.DataFrame): DataFrame da coorte, contendo uma linha por cliente, 
                                   com as colunas 'is_target' e a métrica de interesse.
        coluna_metrica (str): Nome da coluna com os valores agregados (e.g., 'total_amount_mes').

    Returns:
        dict: Dicionário com médias, Lift, P-valor e status de significância.
    """
    
    # 1. Calcula as médias da métrica por grupo
    medias = df_analise.groupby('is_target')[coluna_metrica].mean()
    
    try:
        avg_controle = medias.loc['control']
        avg_teste = medias.loc['target']
    except KeyError as e:
        return {"Erro": f"Grupos 'control' ou 'target' ausentes: {e}. Verifique a coluna 'is_target'."}
    
    # 2. Cálculo do Lift (Teste / Controle - 1)
    lift = (avg_teste / avg_controle) - 1 if avg_controle > 0 else np.nan
    
    # 3. Prepara dados para o Teste t
    data_controle = df_analise[df_analise['is_target'] == 'control'][coluna_metrica]
    data_teste = df_analise[df_analise['is_target'] == 'target'][coluna_metrica]
    
    # Teste t de Student para duas amostras (Welch's t-test, unilateral: Teste > Controle)
    t_stat, p_value = stats.ttest_ind(data_teste, data_controle, equal_var=False, alternative='greater')

    return {
        "Metrica": coluna_metrica,
        "Media_Controle": avg_controle,
        "Media_Teste": avg_teste,
        "Lift": lift,
        "P_Value": p_value,
        "Significativo_05": p_value < 0.05
    }

resultado_amount = calcular_lift_generico(df_janeiro, 'total_amount_mes')

print("--- Resultado do Lift de Amount Total (Valor) ---")
print(f"Média Controle: R$ {resultado_amount['Media_Controle']:.2f}")
print(f"Média Teste:    R$ {resultado_amount['Media_Teste']:.2f}")
print(f"Lift:           {resultado_amount['Lift']:.2%}")
print(f"P-Valor:        {resultado_amount['P_Value']:.4f}")
print(f"Significativo:  {resultado_amount['Significativo_05']}")

###print("--- Resultado do Lift de Frequência (Pedidos) ---")
#print(f"Lift: {resultado_frequencia['Lift']:.2%}")
#print(f"P-Valor: {resultado_frequencia['P_Value']:.4f}")
#print(f"Significativo: {resultado_frequencia['Significativo_05']}")
```
